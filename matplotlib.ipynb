```python
#names=['python','c++','c#','java']
values=[20,30,10,40]
plt.style.use('dark_background')

labels=['python','c++','c#','java']
explodes=[0.05,0.05,0.05,0.05]
plt.pie(values,labels=labels,autopct='%.2f%%',explode=explodes,pctdistance=0.6)
plt.legend(loc='right')
plt.savefig('mypiechart.png',dpi=300)
plt.show()
#plt.savefig('piechart.png')
```

```python
from matplotlib import style
x=np.arange(50)
fig,axs=plt.subplots(2,2)
axs[0,0].plt(x,np.sin(x))
axs[0,0].set_title("sinewave")
axs[0,1].plt(x,np.cos(x))
axs[0,1].set_title("cos")
axs[1,0].plt(x,np.tan(x))
axs[1,0].set_title("tan")
axs[1,1].plt(x,np.log(x))
axs[1,1].set_title("log")
plt.set_title("algebra graph")
plt.show()
```
