```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
points={"blue":[[4,6],[3,4],[4,5],[2,2],[1,1]],"red":[[1,2],[3,5],[6,3],[2,1]]}
new_point=[5,6]
def distancepoint(p,q):
    return np.sqrt(np.sum(np.array(p)-np.array(q))**2)
class kdistance:
    def __init__(self,k=3):
        self.k=k
    def fit(self,points):
        self.points=points
    def predict(self,new_point):
        dis=[]
        for cat in self.points:
            for points in self.points[cat]:
                distance=distancepoint(points,new_point)
                dis.append([distance,cat])
                print(dis)
        categories=[cat[1] for cat in sorted(dis)[:self.k]]
        print(categories)
        result=Counter(categories).most_common(1)[0][0]
        return result
clf=kdistance(k=4)
clf.fit(points)
print(clf.predict(new_point))
ax=plt.subplot()
plt.figure(figsize=(15,9))
ax.grid(True,color='white')
ax.set_facecolor('black')
ax.figure.set_facecolor('blue')
ax.tick_params(axis='x',color='white')
ax.tick_params(axis='y',color='white')
for point in points['blue']:
    ax.scatter(point[0],point[1],color='white')
for point in points['red']:
    ax.scatter(point[0],point[1],color='yellow')
new_class=clf.predict(new_point)
color='grey' if new_point=='red' else 'violet'
ax.scatter(new_point[0],new_point[1],marker='*',color='red')
for point in points["blue"]:
    ax.plot([new_point[0],point[0]],[new_point[1],point[1]],linestyle='--',color='white')
for point in points["red"]:
    ax.plot([new_point[0],point[0]],[new_point[1],point[1]],linestyle='--',color='white')
plt.show()

    
```
