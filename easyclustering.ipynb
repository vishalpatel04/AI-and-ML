```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
player=pd.read_csv("players_20.csv")
player
```

```python
feature=['overall','age','potential','value_eur','wage_eur']
player=player.dropna(subset=feature)
data=player[feature]
data.head()
```

```python
data.isnull().sum()
```

```python
X=data.iloc[:,[3,4]].values
X=((X-X.min())/(X.max()-X.min()))*9+1
X
```

```python
'''ssc=[]
for i in range(1,20):
    kmeans=KMeans(n_clusters=i,n_init=10)
    kmeans.fit(X)
    ssc.append(kmeans.inertia_)
plt.plot(range(1,20),ssc)'''

    
```

```python
kmeans=KMeans(n_clusters=4,n_init=10)
kmeans.fit(X)
```

```python
y=kmeans.fit_predict(X)
y
```

```python
#plt.figure.set_facecolor("black")
plt.scatter(X[y==0,0],X[y==0,1],color='yellow',label='clu1')
plt.scatter(X[y==1,0],X[y==1,1],color='blue',label='clu2')
plt.scatter(X[y==2,0],X[y==2,1],color='green',label='clu3')
plt.scatter(X[y==3,0],X[y==3,1],color='pink',label='clu4')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color='black',marker="*")
plt.legend()
plt.show()
```
